```python
# CONCERTAR O PROBLEMA DO VETOR DE units SER MAIOR QUE O NUMERO DE CAMADAS, O QUE GERA MODELOS REDUNDANTES

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, MultiHeadAttention, LayerNormalization, Attention, Embedding
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.dates as mdates
from keras import regularizers
from skopt import BayesSearchCV
from sklearn.model_selection import TimeSeriesSplit
from keras.callbacks import History
from keras_tuner.tuners import BayesianOptimization
from keras_tuner.engine.hyperparameters import HyperParameters
import tensorflow as tf
from sklearn.metrics import mean_squared_error
from keras.callbacks import EarlyStopping
from keras.callbacks import ReduceLROnPlateau
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)

from tensorflow.keras.callbacks import TensorBoard
import datetime
import os
from pyngrok import ngrok


# Diretório para salvar os logs do TensorBoard
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))

tensorboard_callback = TensorBoard(log_dir=logdir, histogram_freq=1)

# ----------------------------------------------------------------------------------


url = 'https://raw.githubusercontent.com/owid/covid-19-data/cd79eaef83bef59359ecc625777d970a74a8aaa0/public/data/owid-covid-data.csv'
df = pd.read_csv(url, usecols=['location', 'date', 'new_cases'])

# filtra as linhas correspondentes ao Reino Unido
df_uk = df[df['location'] == 'United Kingdom']

# exclui as linhas com valores NaN e zero para new_cases
df_uk = df_uk[(df_uk['new_cases'].notna()) & (df_uk['new_cases'] != 0)]

# converte a coluna de datas para o formato de data
df_uk['date'] = pd.to_datetime(df_uk['date'])

# filtra as datas até 01/02/2024
df_uk = df_uk[df_uk['date'] <= '2024-02-01']

fig, ax = plt.subplots()
ax.plot(pd.to_datetime(df_uk['date']), df_uk['new_cases'])
ax.set_xlabel('Data')
ax.set_ylabel('Casos Confirmados')
ax.set_title('Casos Confirmados de COVID-19 no Reino Unido')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.show()

# -----------------------------------------------
# dividindo os dados
# Definindo as proporções de divisão
train_size = int(len(df_uk) * 0.7) # 70% para treinamento
val_size = int(len(df_uk) * 0.15) # 15% para validação
test_size = len(df_uk) - train_size - val_size # O restante para teste

# Dividindo os dados em conjuntos de treinamento, validação e teste
train_data = df_uk[:train_size]
val_data = df_uk[train_size:train_size+val_size]
test_data = df_uk[train_size+val_size:]
test_data = test_data.drop(['location'], axis=1)
test_data = test_data.drop(['date'], axis=1)


# -------------------------------------------------------
# normalizando os dados

# criando um objeto scaler para normalização
scaler = MinMaxScaler()

# seleciona apenas as colunas numéricas do DataFrame de treinamento
train_data_num = train_data.select_dtypes(include=np.number)

# seleciona apenas as colunas numéricas do DataFrame de treinamento
test_data_num = test_data.select_dtypes(include=np.number)

# seleciona apenas as colunas numéricas do DataFrame de validacao
val_data_num = val_data.select_dtypes(include=np.number)


# normalizando os dados de treinamento
train_norm = scaler.fit_transform(train_data_num)

# normalizando os dados de teste
test_norm = scaler.transform(test_data_num)

# normalizando os dados de validacao
val_norm = scaler.transform(val_data_num)

#--------------------------------------------------------------------------
# preparacao dos dados de entrada e saida

# Define o número de dias a serem usados como entrada e saída
n_input = 30
n_output = 1

# Cria listas vazias para as entradas 'train_X' e saídas 'train_Y'
train_X, train_y = [], []

# Loop pelos dados normalizados de treinamento para criar as entradas e saídas
for i in range(n_input, len(train_norm)):
    train_X.append(train_norm[i-n_input:i, :])
    train_y.append(train_norm[i:i+n_output, 0])

# Converte as listas em arrays numpy
train_X = np.array(train_X)
train_y = np.array(train_y)

# Cria listas vazias para as entradas 'val_X' e saídas 'val_y'
val_X, val_y = [], []

# Loop pelos dados normalizados de validação para criar as entradas e saídas
for i in range(n_input, len(val_norm)):
    val_X.append(val_norm[i-n_input:i, :])
    val_y.append(val_norm[i:i+n_output, 0])

# Converte as listas em arrays numpy
val_X = np.array(val_X)
val_y = np.array(val_y)

# criar dados de teste
test_X, test_y = [], []
for i in range(n_input, len(test_norm)):
    test_X.append(test_norm[i-n_input:i, :])
    test_y.append(test_norm[i:i+n_output, 0])
test_X = np.array(test_X)
test_y = np.array(test_y)


#---------------------------------------------------------------
# construcao do modelo. Foi usado a rede neural do tipo LSTM 'Long Short-Term Memory'

def build_model(hp):
    model = Sequential()

    model.add(LSTM(hp.Int('input_unit',min_value=8,max_value=512,step=8),return_sequences=True, input_shape=(train_X.shape[1],train_X.shape[2])))

    model.add(LSTM(hp.Int('hidden_layer_1',min_value=8,max_value=512,step=8),return_sequences=True))

    model.add(LSTM(hp.Int('hidden_layer_2',min_value=8,max_value=512,step=8)))
    model.add(Dropout(hp.Float('Dropout_rate',min_value=0.1,max_value=0.5,step=0.1)))
    model.add(Dense(1, activation=hp.Choice('dense_activation',values=['relu'],default='relu')))
    model.compile(loss='mean_squared_error', optimizer='adam',metrics = ['mse'])

    return model

# ----------------------------------------------------------------------------------------------------------------------
# primeiro experimento

# def build_model(hp):
#     inputs = keras.Input(shape=(train_X.shape[1], train_X.shape[2]))

#     num_heads=hp.Int('num_heads_1', min_value=2, max_value=24, step=2)

#     # Primeira camada MultiHeadAttention
#     x1 = layers.MultiHeadAttention(
#         num_heads=num_heads,
#         key_dim=int((train_X.shape[1] * train_X.shape[2])/num_heads)
#     )(inputs, inputs)

#     # Skip connections para ambas as camadas
#     res1 = x1 + inputs

#     # Global average pooling
#     x = layers.GlobalAveragePooling1D()(res1)

#     dropouts = layers.Dropout(hp.Float('Dropout_rate',min_value=0.1,max_value=0.5,step=0.1))(x)

#     # Output layer
#     outputs = layers.Dense(1, activation=hp.Choice('dense_activation',values=['relu'],default='relu'))(dropouts)

#     model = keras.Model(inputs, outputs)
#     model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])

#     return model

# ----------------------------------------------------------------------------------------------------------------------
# segundo experimento

# def build_model(hp):
#     inputs = keras.Input(shape=(train_X.shape[1], train_X.shape[2]))

#     num_heads=hp.Int('num_heads_1', min_value=2, max_value=24, step=2)

#     # Primeira camada MultiHeadAttention
#     x1 = layers.MultiHeadAttention(
#         num_heads=num_heads,
#         key_dim=int((train_X.shape[1] * train_X.shape[2])/num_heads)
#     )(inputs, inputs)

#     x1 = layers.Dropout(0)(x1)

#     # Skip connections para ambas as camadas
#     res1 = x1 + inputs

#     # Global average pooling
#     x = layers.GlobalAveragePooling1D(data_format="channels_first")(res1)

#     dropouts = layers.Dropout(hp.Float('Dropout_rate',min_value=0.1,max_value=0.5,step=0.1))(x)

#     # Output layer
#     outputs = layers.Dense(1, activation=hp.Choice('dense_activation',values=['relu'],default='relu'))(dropouts)

#     model = keras.Model(inputs, outputs)
#     model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])

#     return model

# ----------------------------------------------------------------------------------------------------------------------


tuner = BayesianOptimization(
        build_model,
        objective='val_loss',
        max_trials=100,
        executions_per_trial=1,
        overwrite=True
        )


# --------------------------------------------------------------------
# treinando o modelo

early_stopping = EarlyStopping(monitor='val_loss', patience=5)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=0.0001)


tuner.search(
        x=train_X,
        y=train_y,
        epochs=15,
        batch_size=256,
        validation_data=(val_X,val_y),
        callbacks=[early_stopping, reduce_lr, tensorboard_callback]
)

# -----------------------------------------------------------------------
# Rankeia os 10 melhores modelos treinados

best_models = tuner.get_best_models(num_models=10)

# ------------------------------------------------------------------------
# avaliando o modelo

# Criação das listas para armazenar os resultados do teste
test_predictions = []
mse_scores = []


# Obtem o mse do teste dos 10 melhores modelos
for i in range(10):
    # Obtém as previsões do modelo atual para todos os exemplos de teste
    model_predictions = best_models[i].predict(test_X)

    # Armazena as previsões do modelo atual
    test_predictions.append(model_predictions)

    # Calcula o MSE para o modelo atual
    model_mse = mean_squared_error(test_y, model_predictions)
    mse_scores.append(model_mse)


# Imprime os MSE e os hiperparâmetros correspondentes dos 10 melhores modelos
print("Hiperparâmetros dos 10 melhores modelos:")
for i in range(10):
    print(f"Model{i}:")
    print(best_models[i].summary())


print("MSE das previsões dos modelos:")
for i in range(10):
    print(f"Model {i+1}: MSE = {mse_scores[i]}")

# Desnormalizar os casos reais
descaled_test_y = scaler.inverse_transform(test_y)


# Plota os 10 modelos em comparacao com os casos reais
for i in range(10):
    # Obtém as previsões do modelo atual
    model_predictions = test_predictions[i]

    # Desnormaliza as previsões do modelo atual
    descaled_model_predictions = scaler.inverse_transform(model_predictions)

    # Define o intervalo de datas para o eixo x
    dates = range(len(descaled_test_y))

    # Plota os casos diários reais
    plt.plot(dates, descaled_test_y, label='Casos reais', marker='o', markevery=10)

    # Plota as previsões do modelo atual
    plt.plot(dates, descaled_model_predictions, label=f'Modelo {i+1}', marker='s', markevery=10)

    # Configurações do gráfico
    plt.title('Previsões de Casos Diários de COVID-19')
    plt.xlabel('Dias')
    plt.ylabel('Casos Diários')
    plt.legend()

    # Exibe o gráfico
    plt.show()


# Ordena os vetores mse_scores e test_predictions com base nos valores de mse_scores
sorted_indexes = sorted(range(len(mse_scores)), key=lambda k: mse_scores[k])
mse_scores_sorted = [mse_scores[i] for i in sorted_indexes]
test_predictions_sorted = [test_predictions[i] for i in sorted_indexes]

# Desnormaliza as previsões dos 3 melhores modelos
descaled_model_predictions1 = scaler.inverse_transform(test_predictions_sorted[0])
descaled_model_predictions2 = scaler.inverse_transform(test_predictions_sorted[1])
descaled_model_predictions3 = scaler.inverse_transform(test_predictions_sorted[2])

# Define o intervalo de datas para o eixo x
dates = range(len(descaled_test_y))

# Plota os casos diários reais com um marcador de círculo
plt.plot(dates, descaled_test_y, label='Casos reais', marker='o', markevery=10)

# Plota as previsões dos 3 melhores modelos com diferentes marcadores
plt.plot(dates, descaled_model_predictions1, label=f'Modelo {1}', marker='s', markevery=10)  # Quadrado
plt.plot(dates, descaled_model_predictions2, label=f'Modelo {2}', marker='^', markevery=10)  # Triângulo
plt.plot(dates, descaled_model_predictions3, label=f'Modelo {3}', marker='d', markevery=10)  # Diamante

# Configurações do gráfico
plt.title('Previsões de Casos Diários dos 3 melhores modelos')
plt.xlabel('Dias')
plt.ylabel('Casos Diários')
plt.legend()

# Exibe o gráfico
plt.show()


# ----------------------------------------------------------------------------------------

# Inicie o TensorBoard
get_ipython().system_raw(
    'tensorboard --logdir {} --host 0.0.0.0 --port 6006 &'
    .format(logdir)
)

# Inicie o ngrok e crie um túnel para acessar o TensorBoard através da URL gerada
ngrok_url = ngrok.connect(6006)
print('TensorBoard URL:', ngrok_url)
```
